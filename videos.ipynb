```python
import io
import os
import pandas as pd
import requests

tba_key = os.environ['TBA_KEY']
tba = 'https://www.thebluealliance.com/api/v3'
headers = {'X-TBA-Auth-Key': tba_key}
event_key = '2025mimtp'

full_url = f'{tba}/event/{event_key}/matches'
r = requests.get(full_url, headers=headers)
print(f'STATUS CODE: {r.status_code}')
json_data = r.text
df = pd.read_json(io.StringIO(json_data))
```

```python
df
```

```python
for row in df['videos']:
    for video in row:
        if video['type'] == 'youtube':
            # Now make a call to the shell to run yt-dlp
            link = f'https://www.youtube.com/watch?v={video["key"]}'
            os.system(f'yt-dlp {link}')
            break
```
